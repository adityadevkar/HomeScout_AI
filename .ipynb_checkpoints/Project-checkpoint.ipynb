```python
df["price_per_sqft"] = round(df["price"]/df["area"],2)
df.head(15)
```

# Outlier Detection

```python
sns.boxplot(df["price"])
plt.show()
```

```python
df["price_per_sqft"].describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df = remove_pps_outliers(df)
df.shape
```

```python
df.head(10)
```

```python
def locbhk(df,location):
    bhk2 = df[(df.bhk == 2) & (df.location == location)]
    bhk3 = df[(df.bhk == 3) & (df.location == location)]
    plt.figure(figsize=(8,6))
    plt.scatter(bhk2.area,bhk2.price,color="blue",s=50,label="2 bhk")
    plt.scatter(bhk3.area,bhk3.price,color="red",marker="+", s=50,label="3 bhk")
    plt.xlabel("Total sqft")
    plt.ylabel("price")
    plt.title(location)
    plt.legend()

locbhk(df,"Wakad")
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["price_per_sqft"],bins=10)
plt.show()
```

# Predicting for sample

```python
import pandas as pd

def predict_price(location, area, bhk, property_type, model, X):
    x = np.zeros(len(X.columns))

    x[0] = area
    x[1] = bhk

    if location in X.columns:
        loc_index = np.where(X.columns == location)[0][0]
        x[loc_index] = 1

    if property_type in X.columns:
        prop_index = np.where(X.columns == property_type)[0][0]
        x[prop_index] = 1

    #  Convert to DataFrame with column names
    x_df = pd.DataFrame([x], columns=X.columns)

    return model.predict(x_df)[0]
    
price = predict_price(
    location='Baner',
    area=1000,
    bhk=2,
    property_type='Apartment',
    model=lr,
    X=x
)

print(f"Predicted Price: {price:.2f}")
```
